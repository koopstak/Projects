```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
match = pd.read_csv('matches.csv')
match.head(3)
```

```python
delivery = pd.read_csv('deliveries.csv')
delivery.head(3)
```

```python
delivery.shape
```

```python
total_score = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score.shape
```

```python
total_score.head(3)
```

```python
total_score['total_runs']=total_score['total_runs']+1
```

```python
total_score
```

```python
total_score = total_score[total_score['inning']==1]
```

```python
total_score.shape
```

```python
match_df = match.merge(total_score[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df.head(7)
```

```python
match_df['team1'].value_counts()
```

```python
match_df['team1'] = match_df['team1'].replace(['Deccan Chargers'],'Sunrisers Hyderabad')
match_df['team1'] = match_df['team1'].replace(['Delhi Daredevils'],'Delhi Capitals')
```

```python
match_df['team1'].value_counts()
```

```python
match_df['team2'].value_counts()
```

```python
match_df['team2'] = match_df['team2'].replace(['Deccan Chargers'],'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].replace(['Delhi Daredevils'],'Delhi Capitals')
```

```python
match_df['team2'].value_counts()
```

```python
match_df['team1'].unique()
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians',
       'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Delhi Capitals', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals']
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.head(7)
```

```python
match_df.shape
```

```python
match_df['dl_applied'].value_counts()
```

```python
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df.sample(5)
```

```python
match_df.shape
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
delivery_df['current_score'] =  delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left']=126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.sample(3)
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna(0)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x==0 else 1)
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets']  = 10 - wickets
delivery_df.head()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = delivery_df['runs_left']*6/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
delivery_df.head(2)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
final_df.describe()
```

```python
final_df=final_df[final_df['balls_left'] != 0]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
final_df['batting_team']=le.fit_transform(final_df['batting_team'])
final_df['bowling_team']=le.fit_transform(final_df['bowling_team'])
final_df['city']=le.fit_transform(final_df['city'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
pipe = Pipeline(steps=[
    ('step1',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[21]
```
